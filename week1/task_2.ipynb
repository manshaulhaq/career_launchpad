Loading Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler ,StandardScaler
from sklearn.datasets import fetch_openml
```

Loading dataset

```python
titanic = fetch_openml(name='titanic', version=1, as_frame=True)
df = pd.read_csv('titanic.csv')
df.head()
```

Missing Values Fill

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df.head(20)
```

Label Encoding

```python
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df.head(10)
```

```python
df = pd.get_dummies(df, columns=['Embarked','Pclass'])
df.head(10)
```

Normalizing Numerical Features (MinMax Scaler)

```python
scaler = MinMaxScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```

Normalizing Numerical Features (Standard Scaler)

```python
std_scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df.head()
```

Extracting Titles

```python
df['Title'] = df['Name'].str.extract(r',\s*([^\.]*)\s*\.', expand=False)
title_map = {
    "Mr": "Mr", "Miss": "Miss", "Mrs": "Mrs", "Master": "Master",
    "Dr": "Rare", "Rev": "Rare", "Col": "Rare", "Major": "Rare", "Mlle": "Miss",
    "Countess": "Rare", "Ms": "Miss", "Lady": "Rare", "Jonkheer": "Rare",
    "Don": "Rare", "Dona": "Rare", "Mme": "Mrs", "Capt": "Rare", "Sir": "Rare"
}
df['Title'] = df['Title'].map(title_map)
df = pd.get_dummies(df, columns=['Title'], drop_first=True)
df.head(10)
```

Creating Family Sizes

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head(10)
```

```python
df.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
df.head()
```
