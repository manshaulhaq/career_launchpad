3.Create Visualizations

Bar Plots

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Count in Each Class')
plt.xlabel('Pclass')
plt.ylabel('Passenger Count')
plt.legend(title='Survived')
plt.show()
```

Histogram

```python
#df['Survived'].plot(kind='hist')
sns.histplot(data=df, x='Survived', bins=2, discrete=True)
plt.show()
```

Scatter Plots

```python
df.head(25).plot(kind='scatter', x='Survived', y='Name')
```

```python
df.head(50).plot(kind='scatter', x='Survived', y='Age')
```

Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols=df.select_dtypes(include=['int64','float64'])

corr_matrix= numerical_cols.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt='.2f')
plt.show()
```
